```python
#plt.style.use("classic")
plt.figure(figsize=(20,6))
plt.plot (TRX_fil["date"], TRX_fil["close"] , color= "Blue", label= "Valor de cierre")#, linewidth=10 )
plt.legend(loc="upper left")
plt.title("TRX", fontsize=20)
plt.xlabel("Fecha")
plt.tick_params('x', colors="black")
plt.ylabel("Precios")
plt.show()
```
